```python
import pandas as pd

data = pd.read_csv("1976-2020-president.csv")
data.head()
```

```python
data["rank"] = data.groupby(["year", "state"])["candidatevotes"].rank(ascending=False)
data = data[["year", "state", "candidate", "party_simplified", "candidatevotes", "rank"]]
data = data.sort_values(["state", "year", "rank"])
data
```

```python
data.to_csv("election-data.csv", index=False)
```

```python
data = data[(data["state"] == "SOUTH CAROLINA") & (data["rank"] == 1)]
data["previous_winner_party"] = data["party_simplified"].shift(1)
data = data.dropna()
data[data["party_simplified"] != data["previous_winner_party"]]
```
